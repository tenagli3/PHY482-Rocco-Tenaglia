```python
import matplotlib.pyplot as plt   #Import relevant modules
import numpy as np
from numpy import ma
```

## Part 1
<br>
Plot the field of a positive point charge.

```python
q = 1         # Define charge of electron
k = 10**(-5)   # Define scaling constant (Coulomb's constant)


X, Y = np.meshgrid(np.arange(-3.9, 3.9, .2), np.arange(-3.9, 3.9, .2))
EpointchargeX = k*q*X/(X**2+Y**2)**(3/2)
EpointchargeY = k*q*Y/(X**2+Y**2)**(3/2)
```

```python
fig = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X,Y,EpointchargeX,EpointchargeY)

plt.show(fig)
```

## Part 2
<br>
Plot the field of an electric dipole.

```python
x1, y1 = (0,0.05)
x2, y2 = (0,-0.05)

EdipoleX1 = k*q*(X-x1)/((X-x1)**2+(Y-y1)**2)**(3/2)
EdipoleY1 = k*q*(Y-y1)/((X-x1)**2+(Y-y1)**2)**(3/2)
EdipoleX2 = k*q*(X-x2)/((X-x2)**2+(Y-y2)**2)**(3/2)
EdipoleY2 = k*q*(Y-y2)/((X-x2)**2+(Y-y2)**2)**(3/2)

EdipoleX = EdipoleX1 - EdipoleX2
EdipoleY = EdipoleY1 - EdipoleY2
```

```python
fig2 = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X,Y,EdipoleX,EdipoleY)

plt.show(fig)
```

## Part 3
<br>
Plot the field of an electric quadrupole.

```python
x3, y3 = (0.05,0.05)
x4, y4 = (0.05,-0.05)

X, Y = np.meshgrid(np.arange(-3.9, 3.9, .1), np.arange(-3.9, 3.9, .1))
EquadrupoleX1 = k*q*(X-x1)/((X-x1)**2+(Y-y1)**2)**(3/2)
EquadrupoleY1 = k*q*(Y-y1)/((X-x1)**2+(Y-y1)**2)**(3/2)
EquadrupoleX2 = k*q*(X-x2)/((X-x2)**2+(Y-y2)**2)**(3/2)
EquadrupoleY2 = k*q*(Y-y2)/((X-x2)**2+(Y-y2)**2)**(3/2)
EquadrupoleX3 = k*q*(X-x3)/((X-x3)**2+(Y-y3)**2)**(3/2)
EquadrupoleY3 = k*q*(Y-y3)/((X-x3)**2+(Y-y3)**2)**(3/2)
EquadrupoleX4 = k*q*(X-x4)/((X-x4)**2+(Y-y4)**2)**(3/2)
EquadrupoleY4 = k*q*(Y-y4)/((X-x4)**2+(Y-y4)**2)**(3/2)

EquadrupoleX = EquadrupoleX1 - EquadrupoleX2 - EquadrupoleX3 + EquadrupoleX4
EquadrupoleY = EquadrupoleY1 - EquadrupoleY2 - EquadrupoleY3 + EquadrupoleY4
```

```python
fig2 = plt.figure(figsize=(5,4),dpi=150)
plt.quiver(X,Y,EquadrupoleX,EquadrupoleY)

plt.show(fig)
```
