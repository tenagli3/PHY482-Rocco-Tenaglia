```python
import matplotlib.pyplot as plt   #Import relevant modules
import numpy as np
from numpy import ma
```

```python
R = 1
L = 1
w = np.arange(0,100,0.1)
Ratio = R/np.sqrt(R**2 + L**2 * w**2)
```

```python
fig = plt.figure(figsize=(8,8))  ## Create a figure object
plt.plot(w,Ratio)        ## Plot x vs y
plt.xlabel('frequency')            ## Add label to x
plt.ylabel('Vout over Vin')            ## Add label to y
plt.show(fig)
```

This appears to be a low-pass filter, as frequencies below ~10 Hz have significantly higher voltage ratios than higher frequencies.  This implies that $V_{out}$ is larger with respect to $V_{in}$, so the low frequencies have a higher response.

```python
t = np.arange(0,10,0.1)
omega = 1
Vin = 0
for k in range(1,101):
    Vin += 4/np.pi*np.sin((2*k-1)*omega*t)/(2*k-1)
```

```python
fig2 = plt.figure(figsize=(8,8))  ## Create a figure object
plt.plot(t,Vin)        ## Plot x vs y
plt.xlabel('time')            ## Add label to x
plt.ylabel('Vin')            ## Add label to y
plt.show(fig2)
```

```python
Ratio2 = Ratio = (R/np.sqrt(R**2 + L**2 * omega**2))**2      #Redefine ratio as appropriate
Vout = Ratio2 * Vin                                     #Set Vout according to the pre-derived formula

fig3 = plt.figure(figsize=(8,8))                  ## Create a figure object
plt.plot(t,Vout)                                  ## Plot x vs y
plt.xlabel('time')                                ## Add label to x
plt.ylabel('Vout')                                ## Add label to y
plt.show(fig3)
```
