```python
import matplotlib.pyplot as plt   #Import relevant modules
import numpy as np
from numpy import ma
```

```python
V0 = 2.25
R = 0.5
L = 0.05
C = 10

w = np.arange(0,30.1,.1)
```

```python
Imag = V0*R*w*L/np.sqrt((w*L)**2 + (R*(w**2 * L * C - 1))**2)
```

```python
fig = plt.figure(figsize=(8,8))  ## Create a figure object
plt.plot(w,Imag)                 ## Plot x vs y
plt.xlabel('frequency')          ## Add label to x
plt.ylabel('Current Magnitude')      ## Add label to y
plt.show(fig)
```

It appears that the magnitude of the current is 0 both at $w=0$ and at $w \to \infty$.  The interesting point is the spike.  This occurs at ~$w=\frac{1}{2}$ and then dies off at a rate which is inversely dependent on R, L, C.  I have chosen these variables to give a reasonable median graph.  Note that the effect of L is significantly smaller than the effect of C; this is because of a simplification:  <br>
<center>$I_{mag} = \frac{V0R\omega}{\sqrt{\omega^2 + R^2 (\omega^2 C - \frac{1}{L}}}.$</center> <br>
Here we can see that the effect of L is distinctly different from that of C.  The sole effect of the voltage is the height of the spike; I have thus chosen V0 so that the height of the spike is roughly 1.
