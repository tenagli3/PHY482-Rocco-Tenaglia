```python
import matplotlib.pyplot as plt   #Import relevant modules
import numpy as np
from numpy import ma
```

```python
#Define the relevant variables

n_over_n = 2.0
mu1 = 1
mu2 = 1
beta = mu1*n_over_n/mu2

theta = np.arange(0,np.pi/2,0.1)

alpha = np.sqrt(1-(np.sin(theta)/2)**2)/np.cos(theta)

Rratio = (1-beta*alpha)/(1+beta*alpha)
Tratio = 2/(1+beta*alpha)
```

```python
#Now create the graph

fig = plt.figure(figsize=(8,8))  ## Create a figure object
plt.plot(theta,Rratio)        ## Plot x vs y
plt.plot(theta,Tratio)
plt.xlabel('Angle')            ## Add label to x
plt.ylabel('Value of ratio')            ## Add label to y
plt.show(fig)
```

The graph of the T ratio (orange) looks almost identical to that from the previous case.  The R ratio graph (blue) is significantly different in that it slopes downwards instead of upwards (although the starting point is the same and the slope looks similar).  This suggests that the amplitude of the reflected wave is always out of sync with the incident wave.  It also means that there is no Brewster's angle for this case, since Brewster's angle is simply the point where Rratio = 0.

```python
#Recreate figure 9.17

R = Rratio**2              #Define the reflection and transmission coefficients
T = alpha*beta*Tratio**2
Sum = R+T

fig = plt.figure(figsize=(8,8))  ## Create a figure object
plt.plot(theta,R)        ## Plot x vs y
plt.plot(theta,T)
plt.plot(theta,Sum)
plt.xlabel('Angle')            ## Add label to x
plt.ylabel('Value of ratio')            ## Add label to y
plt.show(fig)
```

Here, T is orange, R is blue, and the green line at the top is the sum of the two.  Clearly, the sum is always 1.  Whereas in Griffiths' Figure 9.17 the line for R touched the theta axis, this graph avoids that, again showing the lack of a Brewster's Angle for this case.
